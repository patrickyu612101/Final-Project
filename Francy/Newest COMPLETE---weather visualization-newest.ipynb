```python
# import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
# import csv file
weather = pd.read_csv("daily-weather-in-20-us-cities.csv")
weather.head()
```

```python
# convert the date into year and month
weather['year'] = pd.DatetimeIndex(weather['Date']).year
weather['month'] = pd.DatetimeIndex(weather['Date']).month
weather.head()
```

```python
# rename the columns
weather_renamed = weather.rename(columns={"Austin Temperature (max)":"Austin MaxTemp", "Austin Temperature(min)":"Austin MinTemp", "New York Temperature(max)":"New York MaxTemp", "New York Temperature(min)":"New York MinTemp", "San Francisco Temperature (max)":"San Francisco MaxTemp", "San Francisco Temperature (min)":
                                          "San Francisco MinTemp"})
weather_renamed.head()
```

```python
# filter the data for 2015, 2014, 2013
years = [2015, 2014,2013]
weather_cleaned = weather_renamed[weather.year.isin(years)] 
weather_cleaned.head(50)
```

```python
# concatenate the year and month to colomn "period"
weather_cleaned['period'] = weather_cleaned.year.astype(str).str.cat(weather_cleaned.month.astype(str), sep=' ')
weather_cleaned.head()
```

```python
# data cleaning
austin = weather_cleaned[["Date","Austin Precipitation","Austin MaxTemp","Austin MinTemp","year","month","period"]]
austin.head()
```

```python
# data types
austin.dtypes
```

```python
# convert the data into numeric
austin["Austin MaxTemp"] = pd.to_numeric(austin["Austin MaxTemp"])
austin["Austin MaxTemp"]
austin.to_csv("Austin_price.csv", index=False, header=True)
```

```python
# group the Austin data by column "period"
austin_group =austin.groupby(["period"])
austin_group.head()


```

```python
# data types
austin_group.dtypes
```

```python
# Collect the average max and min temperature in Austin
austin_hightemp = austin_group["Austin MaxTemp"].mean()
austin_lowtemp = austin_group["Austin MinTemp"].mean()
# Collect the years where data was collected
print(austin_lowtemp)
print(austin_hightemp)
```

```python
# plot the chart
Austin_TempHighF, = plt.plot(austin_hightemp, color="red", label="High Temperature Mean")
Austin_TempLowF, =  plt.plot(austin_lowtemp, color="blue", label="Low Temperature Mean")
plt.legend(handles=[Austin_TempHighF, Austin_TempLowF], loc="best")
plt.xlabel("Month")
plt.ylabel("Temperature F")
plt.title("Austin Temperature")
plt.savefig("path/Austin Temperture.jpg")
plt.figure(figsize=(50,50))

plt.show()
```
