```python
# import dependecies
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from matplotlib import style
style.use("ggplot")
# from matplotlib import rcParams
# rcParams['figure.figsize'] = 10, 8
```

```python
# import csv file
home = pd.read_csv("All Homes final fmt.csv")
home.head()
```

```python
# Data cleaning
home_cleaned = home[["Year","Month","NY Home Price/SQFT","SFO Home Price/SQFT","AUS Home Price/SQFT","Unemployment Rate in NY","Unemployment Rate in SFO","Unemployment Rate in AUS","Federal Interest Rate"]]
home_cleaned
```

```python
NY = home_cleaned[["NY Home Price/SQFT","Unemployment Rate in NY","Federal Interest Rate"]]
NY.head()
```

```python
NY.dtypes
```

```python
# get the X,y values for NY
X = NY.iloc[:, 1:3]
y = NY.iloc[:, 0]
print(X)
```

```python
# split the data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# build model 
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# Fitting our model with training data
model.fit(X_train, y_train)

# calculate the model scores
score_train = model.score(X_train, y_train)

score_test = model.score(X_test, y_test)
print(f"R2 Training Score: {score_train}")
print(f"R2 Testing Score: {score_test}")
```

```python
# get the model coefficient and intercept
print('Weight coefficients: ', model.coef_)
print('y-axis intercept: ', model.intercept_) 
```

```python
# using the model to make predictions and create a dataframe with the actual value, predicted value and the error
prediction1 = model.predict(X_test)
pd.DataFrame({"Predicted": prediction1, "Actual": y_test, "Error": prediction1 - y_test})[["Predicted", "Actual", "Error"]]
```

```python

# Plot Residuals
plt.scatter(prediction1, prediction1 - y_test)
plt.hlines(y=0, xmin=prediction1.min(), xmax=prediction1.max())
plt.savefig("path/residual plot for NY Housing Price Predictions.jpg")
plt.show()
```

```python
# group the housing data by year
home_grouped = home_cleaned.groupby(["Year"])
home_grouped.head()
```

```python
home_grouped.count().head()
```

```python
# calculate the average of the price per SQFT in three cities
austin_housing_year = home_grouped["AUS Home Price/SQFT"].mean()
SFO_housing_year  = home_grouped["SFO Home Price/SQFT"].mean()
NY_housing_year  = home_grouped["NY Home Price/SQFT"].mean()
# Collect the years where data was collected
print(austin_housing_year)
print(SFO_housing_year)
print(NY_housing_year)
```

```python
# plot the chart
Austin_housing, = plt.plot(austin_housing_year, color="red", label="Austin Housing Price/Sqft")
SFO_housing, = plt.plot(SFO_housing_year, color="blue", label="San Francisco Housing Price/Sqft")
NY_housing, = plt.plot(NY_housing_year, color="yellow", label="New York Housing Price/Sqft")
plt.legend(handles=[Austin_housing, SFO_housing,NY_housing], loc="best")
plt.xlabel("Years from 2010~2018")
plt.ylabel("Housing Price/Sqft")
plt.title("Housing Price/Sqft in Austin, NY, SFO")
plt.savefig("path/housing price comparison for three cities.jpg")
plt.figure(figsize=(50,50))

plt.show()
```
